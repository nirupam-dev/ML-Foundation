```python
import pandas as pd

data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", "Missing", "Sameer"],
    "Age": [22, 23, None, 22, 25, 24, 26],
    "Marks": [85, None, 88, 90, 56, 72, None],
    "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Analyze the logic for Exercise 5
# Technical Q1: Check why 'Marks' column has floats
print(df['Marks'])
print(df['Marks'].dtype)

# Technical Q2: Demonstrate inplace=True risk when dropping missing values
df_copy = df.copy()
df_copy.dropna(inplace=True)
print(df_copy)
print(df)

# Business Q: Check for missing or inconsistent values in 'Name' column
print(df['Name'])
print(df['Name'].isnull().sum())
```
