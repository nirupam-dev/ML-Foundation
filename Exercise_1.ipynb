```python
### Answers to Task 5 Questions

# - **Why are `Marks` values floats?**  Because the column contains `NaN` values, pandas uses a floating-point dtype to accommodate both integers and missing values. When the missing entries are dropped or filled, you could convert to an integer type, but the default numeric type for mixed integer/NaN data is `float`.

# - **Risk of using `inplace=True` when dropping missing values:**  Using `inplace=True` modifies the original `DataFrame` directly. This can lead to loss of the original data if you later realize you needed the unmodified version, and it makes debugging or reverting changes harder. It's safer to assign the result to a new variable (like `cleaned_df`) or reassign explicitly to the same name.

# - **Feedback for the admin team about the `Name` column:**  There is at least one missing name (`None`/`NaN`) in the dataset. The admin team should ensure a valid name is provided for every student and prevent records with missing names from being entered.

```
